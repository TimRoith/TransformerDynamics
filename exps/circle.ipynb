```python
%load_ext autoreload
%autoreload 2
import tfdy
import tfdy.plotting as tplt
import torch
import matplotlib.pyplot as plt
```

## Set up problem Matrix

```python
d1 = 0.5
d2 = 1
n = 300
D = 1.* torch.tensor([[d1,0], [0, d2]])
```

```python
optim = tfdy.optim_diracs.OptimDiracs2D(D=D)
x = optim.optimize(n = n, max_it=1000, phi=tfdy.utils.init_phi_2D(n), opt_kwargs = {'lr':0.01})
```

```python
fig, ax = plt.subplots()
tplt.vis_optim(ax, optim)
```
