```python
%load_ext autoreload
%autoreload 2
import tfdy
import tfdy.plotting as tplt
import torch
```

```python
d1 = 0.5
d2 = 1
d3 = 1
n = 300
D = torch.diag(torch.tensor([d1, d2, d3]))
```

```python
optim = tfdy.optim_diracs.OptimDiracs(D=D)
x = optim.optimize(n = n, max_it=1000, phi=None, opt_kwargs = {'lr':0.01})
```

```python
%matplotlib ipympl
import matplotlib.pyplot as plt
%matplotlib ipympl
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2])
```
